```python
# Write a program that reads a file, breaks each line into words, strips whitespace and punctuation from the words, 
#and converts them to lowercase.
#Hint: The string module provides strings named whitespace, which contains space, tab, newline, etc., 
#punctuation which contains the punctuation characters and printable which contains printable characters.
#import string
#print (string.punctuation)
#!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~
#Also, you might consider using the string methods strip and replace

import string

specialCharacters = string.punctuation
print(specialCharacters)
fin = open('mybook.txt', encoding="utf-8")# opened the downloaded textbook in 8 bit format
for line in fin:
    for word in line.split():
        word = word.strip(specialCharacters)#removing the special charcters of the words 
        word = word.lower()#converting the words into lower case 
        print(word)
```

```python
#Task 2:
#Modify your program from the previous exercise to read the book you downloaded and process the rest of the words as before.
#A. Then modify the program to count the total number of words in the book, and the number of times each word is used.
#B. print the 20 most frequently-used words in the book.
#C. Print the number of different words used in the book
#Optional: write a function that takes in any number of books and produce the book with the greatest number of words

import string

specialCharacters = string.punctuation
fin = open('mybook.txt', encoding="utf-8")
dict = {}
for line in fin:
    for word in line.split():
        word = word.strip(specialCharacters)
        word = word.lower()
        if word not in dict: # condition for adding the words list into dictionary.
            dict[word] = 1
        else:
            dict[word] = dict[word] + 1
print('total number of words') 
print(str(sum(dict.values()))) # printing the count of words by adding the values of words.
```
